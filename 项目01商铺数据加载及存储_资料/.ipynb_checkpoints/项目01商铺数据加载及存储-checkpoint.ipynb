```python
'''
【项目01】  商铺数据加载及存储

作业要求：
1、成功读取“商铺数据.csv”文件
2、解析数据，存成列表字典格式：[{'var1':value1,'var2':value2,'var3':values,...},...,{}]
3、数据清洗：
① comment，price两个字段清洗成数字
② 清除字段缺失的数据
③ commentlist拆分成三个字段，并且清洗成数字
4、结果存为.pkl文件

'''

import os,pickle
import csv
path = r'C:\Users\rjw\Desktop\dest\项目01商铺数据加载及存储_资料'
os.chdir(path)
 
def commentFilter(s):
    if '条点评' in s:
        return(int(s.split(' ')[0]))
    else:
        return "缺失数据"
     
def priceFilter(s):
    if '￥' in s:
        return(int(s.split('￥')[1]))
    else:
        return "缺失数据"
     
def clFilter(s):
    if len(s) == 3:
        quality = float(s[0][2:])
        environment = float(s[1][2:])
        service = float(s[2][2:])
        return [quality, environment, service]
    else:
        return "缺失数据"
 
 
datalst = []
 
with open('商铺数据.csv', newline='', encoding='utf-8') as f:
    reader = csv.reader(f)
    next(reader)
    for row in reader:
        #print(commentFilter(row[2]))
        #print(priceFilter(row[4]))
        #print(clFilter(row[-1].split('                                '))[0])
        classify = row[0]
        name = row[1]
        commentNum = commentFilter(row[2])
        star = row[3]
        price = priceFilter(row[4])
        address = row[5]
        quality = clFilter(row[-1].split('                                '))[0]
        env = clFilter(row[-1].split('                                '))[1]
        service = clFilter(row[-1].split('                                '))[2]
        if "缺失数据" not in [commentNum, price, quality, env, service]:
            temp = [['classify',classify],['name',name],
                    ['commentNum',commentNum],['star',star],
                   ['price',price],['address',address],
                   ['quality',quality],['env',env],['service',service]]
            datalst.append(dict(temp))
        else:
            continue
 
pic = open('C:/Users/rjw/Desktop/data.pkl','wb')
pickle.dump(datalst,pic)
pic.close()
```
